```python
def find_adjacent_positions(x, y):
    adjacent_positions = [
        (x + 1, y),     
        (x - 1, y),     
        (x, y + 1),     
        (x, y - 1),     
        (x + 1, y + 1), 
        (x - 1, y + 1), 
        (x + 1, y - 1), 
        (x - 1, y - 1) 
    ]
    return adjacent_positions
```

```python
grid_input = open("input.txt").read().strip().split("\n")
grid = [[x for x in line] for line in grid_input]

padded_grid = [["Y"] * (len(grid[0]) + 2)]
for line in grid:
    padded_grid.append(["Y"] + line + ["Y"])
padded_grid.append(["Y"] * (len(grid[0]) + 2))

tot = 0

for l in range(len(padded_grid)):
    for c in range(len(padded_grid[l])):
        if padded_grid[l][c] == "@":
            adjacent = find_adjacent_positions(c, l)
            counter = 0
            for pos in adjacent:
                adj_x, adj_y = pos
                if padded_grid[adj_y][adj_x] == "@":
                    counter += 1
            if counter < 4:
                tot += 1
print(tot)
```

```python
tot = 0

while True:
    prev_tot = tot
    for l in range(len(padded_grid)):
        for c in range(len(padded_grid[l])):
            if padded_grid[l][c] == "@":
                adjacent = find_adjacent_positions(c, l)
                counter = 0
                for pos in adjacent:
                    adj_x, adj_y = pos
                    if padded_grid[adj_y][adj_x] == "@":
                        counter += 1
                if counter < 4:
                    tot += 1
                    padded_grid[l][c] = "x"
    if tot == prev_tot:
        break

print(tot)
```
