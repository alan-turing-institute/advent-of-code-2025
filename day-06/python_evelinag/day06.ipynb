```python
test_input = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """

with open('input.txt','r') as f:
    full_input = f.read()

text = full_input
homework = [ l.split() for l in text.splitlines() ]
print(homework)
```

```python
import math

numbers = homework[:-1]
operations = homework[-1]
print(numbers, operations)

k = len(operations)
n = 0
for i in range(k):
  ns = [ int(values[i]) for values in numbers ]
  if operations[i] == '+':
    n += sum(ns)
  else:
    n += math.prod(ns)

print(n)    
```

```python
length = len(text.splitlines()[0])
gap_columns = set(range(length))
raw_lines = text.splitlines()[:-1]
print(gap_columns)

for line in raw_lines:
  full = set()
  for i, s in enumerate(line):
    if s != ' ': full.add(i)
  gap_columns -= full

print(gap_columns)
```

```python
operation_idx = 0
total = 0
partial_total = 0

for i in range(length):
  if i in gap_columns:
    operation_idx += 1
    total += partial_total
    partial_total = 0 if operations[operation_idx] == '+' else 1
  else:
    number = int(''.join([line[i] for line in raw_lines]))
    if operations[operation_idx] == '+':
      partial_total += number
    else:
      partial_total = partial_total * number 

total += partial_total
print(total)
```

```python
import pandas as pd

lines = text.splitlines()
df = pd.DataFrame([list(line.ljust(length)) for line in lines[:-1]])
operations = lines[-1].split()
print(df)
```

```python
gap_columns = df.columns[df.eq(' ').all()]
print(gap_columns)

# Group columns by operation, split at gaps
column_groups = []
current_group = []
for col in df.columns:
    if col in gap_columns:
        if current_group:
            column_groups.append(current_group)
            current_group = []
    else:
        current_group.append(col)
if current_group:
    column_groups.append(current_group)

# Process each group
total = 0
for operation, cols in zip(operations, column_groups):
    numbers = df[cols].apply(lambda row: int(''.join(row)), axis = 1)
    print(numbers)
    if operation == '+':
        total += numbers.sum()
    else:
        total += numbers.prod()

print(total)
```
