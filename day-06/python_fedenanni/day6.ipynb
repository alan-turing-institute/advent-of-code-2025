```python
data = open('test.txt', 'r').read().splitlines()
data = [line.split() for line in data]
operations = data[-1]

tot = 0

for o in range(len(operations)):
    operation = operations[o]
    numbers = [int(data[i][o]) for i in range(len(data)-1)]
    if operation == '+':
        tot += sum(numbers)
    elif operation == '*':
        prod = 1
        for n in numbers:
            prod *= n
        tot += prod
print(tot)
```

```python
# make a loop on operations from the end to the start
data = open('input.txt', 'r').read().splitlines()
operation_line = data[-1]
where_to_split = []
for i in range(1, len(operation_line)):
    if operation_line[i] == '+' or operation_line[i] == '*':
        where_to_split.append(i-1)

new_data = []
start = 0
for split in where_to_split:
    new_data.append([line[start:split] for line in data])
    start = split+1
new_data.append([line[start:] for line in data])

operations = [line[-1].strip() for line in new_data]

# remove operation symbols from new_data
new_data = [line[:-1] for line in new_data]

tot = 0

for o in range(len(operations)-1, -1, -1):
    operation = operations[o]
    numbers = [new_data[o][i] for i in range(len(new_data[o]))]
    max_len = max(len(n) for n in numbers)
    if operation == '+':
        numbers = [number for number in numbers]
        new_numbers = []
        for i in range(max_len):
            new_number = ''
            for number in numbers:
                if i < len(number):
                    new_number += number[i]
                else:
                    new_number += ''
            new_numbers.append(new_number)
        new_numbers = [int(n) for n in new_numbers]
        tot += sum(new_numbers)
        
    elif operation == '*':
        numbers = [number for number in numbers]
        new_numbers = []
        for i in range(max_len):
            new_number = ''
            for number in numbers:
                if number == '':
                    continue
                new_number += number[i]
            new_numbers.append(new_number)
            
        new_numbers = [int(n) for n in new_numbers]
        prod = 1
        for n in new_numbers:
            prod *= n
        tot += prod
print (tot)
```
