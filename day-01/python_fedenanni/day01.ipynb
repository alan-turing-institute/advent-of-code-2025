```python
input = open('input.txt', 'r').read().strip().split('\n')

start = 50

how_many_zeros = 0

for line in input:
    directions = line[0]
    distance = int(line[1:].strip())
    if directions == 'L':
        move = start - distance
    elif directions == 'R':
        move = start + distance
    
    move = move % 100 
    
    start = move
    if start == 0:
        how_many_zeros += 1
    
print(how_many_zeros)
```
