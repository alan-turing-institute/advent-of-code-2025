```python

from itertools import accumulate

with open('input.txt','r') as f:
    input_moves = f.readlines()



def parse(rotation: str) -> int:
    value = int(rotation[1:])
    return -value if rotation[0] == 'L' else value

moves = [parse(x) for x in input_moves]



positions = list(accumulate(moves, lambda position, rotation: (position + rotation)%100, initial=50))

part1 = sum(x == 0 for x in positions)
print(part1)
```

```python
# numpy alternative for part 1
import numpy as np

positions = np.cumsum([50] + moves) % 100
np.sum(positions == 0)
```

```python
zeros = 0

for x, m in zip(positions[:-1], moves):
    visited = list(range(x, x + m, -1 if m < 0 else 1))
    zeros += sum(v % 100 == 0 for v in visited)


print(zeros)
```
