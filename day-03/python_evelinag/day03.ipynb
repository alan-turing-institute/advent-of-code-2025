```python
test_input = """987654321111111
811111111111119
234234234234278
818181911112111""".splitlines()

with open('input.txt','r') as f:
    full_input = f.read().splitlines()

#packs = test_input
packs = full_input
```

```python
def joltage(battery, k):
    numbers = [int(j) for j in battery]

    result = []
    for n in range(k-1,-1,-1):
        idx = max(range(len(numbers)-n), key=lambda i: numbers[i])
        result.append(str(numbers[idx]))
        numbers = numbers[idx+1:]

    return int(''.join(result))


part1 = sum(joltage(b, 2) for b in packs)
part2 = sum(joltage(b, 12) for b in packs)

print(part1)
print(part2)
```
