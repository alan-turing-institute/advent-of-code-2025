```python
batteries = [[int(num) for num in x] for x in open('input.txt', 'r').read().splitlines()]

tot = 0
for battery in batteries:
    max_index = battery.index(max(battery[:-1]))
    max_second_index = battery.index(max(battery[max_index+1:]))
    tot += battery[max_index] * 10 + battery[max_second_index]
print(tot)
```

```python
tot = 0
for battery in batteries:
    tot_voltage = 12
    found_values = []
    while tot_voltage > 0:
        if len(battery) == 1 :
            found_values.extend(battery)
            tot_voltage = 0
            continue
        if tot_voltage == 1:
            found_values.append(max(battery))
            tot_voltage = 0
            continue
        max_index = battery.index(max(battery[:-tot_voltage+1]))
        found_values.append(battery[max_index])
        tot_voltage -= 1
        battery = battery[max_index+1:]
    number = "".join([str(x) for x in found_values])
    tot += int(number)
print(tot)
```
