```python
ranges = open('input.txt', 'r').read().strip().split(',')
tot = 0
for r in ranges:
    first_id, last_id = map(int, r.split('-'))
    for number in range(int(first_id), int(last_id) + 1):
        first_half, second_half = str(number)[:len(str(number))//2], str(number)[len(str(number))//2:]
        if first_half == second_half:
            tot += number
print(tot)
```

```python
def find_all_possible_equal_groups(number):
    num_digits = len(str(number))

    valid_group_lengths = [k for k in range(2, num_digits + 1) if num_digits % k == 0]

    return valid_group_lengths
```

```python
tot = 0
for r in ranges:
    first_id, last_id = map(int, r.split('-'))
    for number in range(int(first_id), int(last_id) + 1):
        valid_group_lengths = find_all_possible_equal_groups(number)
        str_number = str(number)
        for group_length in valid_group_lengths:
            group_size = len(str_number) // group_length
            groups = [str_number[i * group_size:(i + 1) * group_size] for i in range(group_length)]
            if all(g == groups[0] for g in groups):
                tot += number
                break

print(tot)
```
