```python

with open('input.txt') as f:
    input = f.read().split(',')



def count_invalid(id_range: str) -> int:
    score = 0
    lower, upper = map(int,id_range.split('-'))

    for n in range(lower, upper+1):
        s = str(n)
        l = len(s)
        if l % 2 == 0 and s[:l//2] == s[l//2:]:
            score += n
    
    return score

sum(map(count_invalid, input))    
```

```python
def count_invalid2(id_range: str) -> int:
    ids = []
    lower, upper = map(int,id_range.split('-'))

    for n in range(lower, upper+1):
        s = str(n)
        l = len(s)

        for k in range(1,l//2+1):
            chunks = set([s[i:i+k] for i in range(0, len(s), k)])
            if len(chunks) == 1:
                ids.append(n)

    return sum(set(ids))

sum(map(count_invalid2, input))
```
