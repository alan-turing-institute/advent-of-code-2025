```python
input = open('input.txt', 'r').readlines()

lines = [list(line.strip()) for line in input]

starts = [lines[0].index('S')]
next_line = 1

tot = 0

while next_line < len(lines):
    new_start = []
    for start in starts:
        if lines[next_line][start] == '.':
            new_start.append(start)
            lines[next_line][start] = '|'
        elif lines[next_line][start] == '^':
            tot += 1
            new_start.append(start - 1)
            new_start.append(start + 1)
            lines[next_line][start - 1] = '|'
            lines[next_line][start + 1] = '|'
    starts = new_start
    next_line += 1

print(tot)
```

```python
start = lines[0].index('S')
start_row = 0

rows = len(lines)
cols = len(lines[0])

cur = [0] * cols
cur[start] = 1

for r in range(start_row, rows - 1):
    nxt = [0] * cols
    for c in range(cols):
        count = cur[c]
        if count == 0:
            continue
        cell = lines[r+1][c]
        if cell == '^':
            if c - 1 >= 0:
                nxt[c - 1] += count
            if c + 1 < cols:
                nxt[c + 1] += count
        else:
            nxt[c] += count
    cur = nxt
sum(cur)
```
