```python
test_input = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............""".splitlines()

with open('input.txt','r') as f:
    full_input = f.read().splitlines()

manifold = full_input
```

```python


def tachyon(state, level: str) -> set:
    beams, count = state
    if not beams:
        beams.add(level.index('S'))
    else:
        for i, x in enumerate(level):
            if x == '^':
                if i in beams:
                    count += 1
                    beams.remove(i)
                    beams.add(i+1)
                    beams.add(i-1)
    return (beams, count)

from functools import reduce

_, count = reduce(tachyon, manifold, (set(), 0))
print(count)
```

```python
def qtachyon(counter, level: str) -> set:
    #print(' '.join([str(x) for x in counter]).replace('0', '.'))
    updated_counter = [0] * len(level)
    if '^' in level:
        for i, c in enumerate(counter):
            if counter[i] > 0:
                if level[i] == '^':
                    updated_counter[i-1] += counter[i] 
                    updated_counter[i+1] += counter[i] 
                else:
                    updated_counter[i] += counter[i]
        return updated_counter
    else:
        return counter

from functools import reduce

counter = [0] * len(manifold[0])
counter[manifold[0].index('S')] = 1

final = reduce(qtachyon, manifold[1:], counter)
print(sum(final))
```
